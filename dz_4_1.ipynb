**Задание 2**

С помощью функции linspace из библиотеки Numpy создайте массив t из 51 числа от 0 до 10 включительно.

Создайте массив Numpy под названием f, содержащий косинусы элементов массива t.

Постройте линейную диаграмму, используя массив t для координат по горизонтали,а массив f - для координат по вертикали. Линия графика должна быть зеленого цвета.

Выведите название диаграммы - 'График f(t)'. Также добавьте названия для горизонтальной оси - 'Значения t' и для вертикальной - 'Значения f'.

Ограничьте график по оси x значениями 0.5 и 9.5, а по оси y - значениями -2.5 и 2.5.

```python
# С помощью функции linspace из библиотеки Numpy создайте массив t из 51 числа от 0 до 10 включительно.

t = np.linspace(0, 10, 51)
# посмотреть количество элементов и убедиться, что их 51
t.size
```

```python
# Создайте массив Numpy под названием f, содержащий косинусы элементов массива t

f = np.cos(t)

# Постройте линейную диаграмму, используя массив t для координат по горизонтали,а массив 
# f - для координат по вертикали. Линия графика должна быть зеленого цвета.

plt.plot(t, f, color='green')

# Выведите название диаграммы - 'График f(t)'. 
# Также добавьте названия для горизонтальной оси - 'Значения t' и для вертикальной - 'Значения f'.

plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')

# Ограничьте график по оси x значениями 0.5 и 9.5, а по оси y - значениями -2.5 и 2.5.
plt.axis([0.5, 9.5, -2.5, 2.5])

plt.show()
```

***Задание 3**

С помощью функции linspace библиотеки Numpy создайте массив x из 51 числа от -3 до 3 включительно.

Создайте массивы y1, y2, y3, y4 по следующим формулам:

    y1 = x**2
    y2 = 2 * x + 0.5
    y3 = -3 * x - 1.5
    y4 = sin(x)

Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с названием fig и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных графика в сетке, состоящей из двух строк и двух столбцов. В каждом графике массив x используется для координат по горизонтали.В левом верхнем графике для координат по вертикали используйте y1,в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4.Дайте название графикам: 'График y1', 'График y2' и т.д.

Для графика в левом верхнем углу установите границы по оси x от -5 до 5.

Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.

Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.

```python
# С помощью функции linspace библиотеки Numpy создайте массив x из 51 числа от -3 до 3 включительно.

x = np.linspace(-3, 3, 51)
# посмотреть количество элементов и убедиться, что их 51
x.size
```

```python
# Создайте массивы y1, y2, y3, y4 по следующим формулам:
## y1 = x**2
## y2 = 2 * x + 0.5
## y3 = -3 * x - 1.5
## y4 = sin(x)

y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)

# Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с названием fig 
# и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных графика в сетке, состоящей из двух строк 
# и двух столбцов. В каждом графике массив x используется для координат по горизонтали.В левом верхнем графике 
# для координат по вертикали используйте y1,в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4.
# Дайте название графикам: 'График y1', 'График y2' и т.д.

fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_title('График y1')
ax2.plot(x, y2)
ax2.set_title('График y2')
ax3.plot(x, y3)
ax3.set_title('График y3')
ax4.plot(x, y4)
ax4.set_title('График y4')

# Для графика в левом верхнем углу установите границы по оси x от -5 до 5.

ax1.set_xlim([-5, 5])

# Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.

fig.set_size_inches(8, 6)

# Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.

plt.subplots_adjust(hspace=0.3, wspace=0.3)
```

```python
# а также используйте для графиков стиль “fivethirtyeight”
# прописываем этот стиль

plt.style.use('fivethirtyeight')
```

```python
# Посчитайте с помощью метода value_counts количество наблюдений для каждого значения целевой переменной Class 

class_count = df['Class'].value_counts()
class_count
```

```python
# и примените к полученным данным метод plot, чтобы построить столбчатую диаграмму

class_count.plot(kind='bar', color='pink')
plt.show()
```

```python
# Затем постройте такую же диаграмму, используя логарифмический масштаб.

class_count.plot(kind='bar', logy=True, color='pink')
plt.show()
```

```python
# На следующем графике постройте две гистограммы по значениям признака V1 - одну для мошеннических транзакций (Class равен 1) 
# и другую - для обычных (Class равен 0). Подберите значение аргумента density так, чтобы по вертикали графика было 
# расположено не число наблюдений, а плотность распределения. Число бинов должно равняться 20 для обеих гистограмм, а 
# коэффициент alpha сделайте равным 0.5, чтобы гистограммы были полупрозрачными и не загораживали друг друга. Создайте легенду 
# с двумя значениями: “Class 0” и “Class 1”. Гистограмма обычных транзакций должна быть серого цвета, а мошеннических - 
# красного. Горизонтальной оси дайте название “V1”.

plt.hist(df.loc[df['Class'] == 1, ['V1']], density=True, bins=20, alpha=0.5, label='Class 1', color='#B22222')
plt.hist(df.loc[df['Class'] == 0, ['V1']], density=True, bins=20, alpha=0.5, label='Class 0', color='#696969')

plt.xlabel('V1')
plt.legend()
plt.show()
```
